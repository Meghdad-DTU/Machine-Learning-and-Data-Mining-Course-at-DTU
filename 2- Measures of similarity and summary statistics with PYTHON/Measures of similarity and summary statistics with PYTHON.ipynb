#### Method 2

```python
file = open("text doc/textDocs.txt")
lines = list(file)
textDocs=[line.strip() for line in lines]
        
#docName=["doc%d"%(i+1) for i in range(0,5)] 
documnets=[]
for textDoc in textDocs:   
    if len(textDoc)>0:
        documnets.append(textDoc)
documnets
```

```python
# Scikit Learn
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

# Create the Document Term Matrix
count_vectorizer = CountVectorizer(stop_words='english')
count_vectorizer = CountVectorizer()
sparse_matrix = count_vectorizer.fit_transform(documnets)

# OPTIONAL: Convert Sparse Matrix to Pandas Dataframe if you want to see the word frequencies.
doc_term_matrix = sparse_matrix.todense()
df = pd.DataFrame(doc_term_matrix, 
                  columns=count_vectorizer.get_feature_names(), 
                  index=['doc1', 'doc2', 'doc3','doc4', 'doc5'])
print(len(df.columns))
df
```

```python
f = open("text doc/stopWords.txt")
lines = list(f)
stopWords=[line.strip() for line in lines]

common=[]
for x in df.columns:
    for stopWord in stopWords:
        if x==stopWord:
            common.append(x)
df=df.drop(common,axis=1)
df.shape[1]
df
```
